```python
def get_dataset(df_results, df_tour, year):
    answers = []
    members = []
    tournaments = []
    qty = []
    round_qty = []
    numbers_qt = []
    teams = []
    places = []
    quest_id = []
    for tournament_id in df_tour:
        if not df_tour[tournament_id]['dateStart'].split("-")[0] == year:
            continue
        for el in df_results[tournament_id]:
            rounds = []
            for key, value in df_tour[tournament_id]["questionQty"].items():
                rounds.extend([key] * value)
            answers_to_quest = el.get("mask", [])
            #print(rounds, answers_to_quest)
            num_of_questions = len(answers_to_quest) if answers_to_quest is not None else 0
            for member in [member["player"]["id"] for member in el["teamMembers"]]:
                for i in range(num_of_questions):
                    #print(round_num, i)
                    #print(answers_to_quest, num_of_questions, num_of_rounds, tournament_id)
                    if answers_to_quest[i] == '1' or answers_to_quest[i] == '0':
                      qty.append(df_tour[tournament_id]["questionQty"])
                      tournaments.append(tournament_id)
                      members.append(member)
                      teams.append(el['team']['id'])
                      quest_id.append(str(tournament_id)+'_'+str(i))
                      answers.append(int(answers_to_quest[i]))
                      numbers_qt.append(i + 1)
                      round_qty.append(int(rounds[min(i, len(rounds) - 1)]))
                      places.append(el['position'])
    dfs = pd.DataFrame({"answer": answers, "player_id": members, "tournament_id": tournaments, 'team_id': teams, 'position': places, 'question_id': quest_id})
    dfs = dfs[~dfs.answer.isna()]
    return dfs
#dfs.head()
train_dataset = get_dataset(df_results, df_tour, "2019")
test_dataset = get_dataset(df_results, df_tour, "2020")
```

## Linear model

качестве baseline модели построим линейную регрессию, которая будет обучать рейтинг-лист игроков.
для каждого игрока есть запись с ответом на вопрос, и сложностью вопроса. Сделаем one-hot по id участика и id вопроса, таргет - бинарный ответ на вопрос
train - турниры до 2019 года, test - после 

```python
train_dataset['player_id'] = train_dataset['player_id'].astype('category')
train_dataset['question_id'] = train_dataset['question_id'].astype('category')
X_train = pd.get_dummies(train_dataset[['player_id', 'question_id']], sparse = True)
y_train = train_dataset['answer']
X_train
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
def get_rating(model, data, dict_res):
  cols = data.columns
  coef_rating = model.coef_
  for i in range(59101):
      ids = int(cols[i].split('_')[-1])
      dict_res[ids] = coef_rating[i]
```

```python
COEF_MODEL_DICT = {}
get_rating(model, X_train, COEF_MODEL_DICT)
```
